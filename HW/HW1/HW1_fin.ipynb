# Numpy Tutorial
Numpy is a high performance framework for operations on multi dimensional arrays.

```python
import numpy as np

# Create a simple (2, 3) numpy array
# [[ 1  2  3]
#  [ 4  5  6]]
a = np.array([[1,2,3], [4,5,6]])
print('a='); print(a)

# dtype可以印出資料型別
print('Data type:', a.dtype)

# shape 可以印出向量的形狀(列 & 行)
print('Shape of data:', a.shape)
```

```python
# 切割運算 
# Let us take the first two columns from the original array and save it in a new one
# 逗點是用來隔開不同維度的
b = a[:,:2]
print('b='); print(b)
print('Data type:', b.dtype)
print('Shape of data:', b.shape)
```

```python
# Select elements which satisfy a condition
# Using numpy array makes such a selection trivial
# 利用遮罩(mask)來取得部分的值
print('a='); print(a)
print('')

mask = a > 1
print('mask='); print(mask)
print('')

new_array = a[mask]
print('a[mask]='); print(new_array)
```

```python
# Do the same thing in a single step
c = a[a > 1]
print(np.all(new_array == c)) # 若兩矩陣完全吻合則 np.all() 會輸出 True，反之則 False
```

```python
# 製作值都一樣的矩陣
print('all-zero (2, 3) array:'); print(np.zeros((2, 3)))
print('')

print('all-one (2, 3) array:'); print(np.ones((2, 3)))
print('')

# 所有值都是指定的
print('all filled with a constant (2, 3) array:'); print(np.full((2, 3), 2.5))
print('')

# 隨機
print('filled with random values (2, 3) array:'); print(np.random.random((2, 3)))
```

```python
# Mathematical operations
x = np.array([[1,2],[3,4]])
y = np.array([[5,6],[7,8]])

# Elementwise sum; both produce the array
# [[ 6  8]
#  [10 12]]
print('x + y='); print(x + y)
print('numpy.add(x, y)='); print(np.add(x, y))
print('')

# Elementwise difference; both produce the array
# [[-4 -4]
#  [-4 -4]]
print('x - y='); print(x - y)
print('numpy.subtract(x, y)=:'); print(np.subtract(x, y))
print('')

# Elementwise product; both produce the array
# [[ 5 12]
#  [21 32]]
print('x * y='); print(x * y)
print('np.multiply(x, y)='); print(np.multiply(x, y))
print('')

# Elementwise division; both produce the array
# [[0.2      0.33333333]
#  [0.42857143     0.5]]
print('x / y='); print(x / y)
print('np.divide(x, y)='); print(np.divide(x, y))
```

```python
print('x='); print(x)
print('')

# Element wise operations
# Elementwise square root; produces the array
# [[1.       1.41421356]
#  [1.73205081      2.]]
# 對矩陣開根號
print('np.sqrt(x)='); print(np.sqrt(x))
print('')

# Element wise exponential operation
# [[ 2.71828183  7.3890561]
#  [20.08553692  54.59815003]]
# 取 e 的次方
print('np.exp(x)='); print(np.exp(x))
```

Broadcasting is a powerful mechanism that allows numpy to work with arrays of different shapes when performing arithmetic operations. Frequently we have a smaller array and a larger array, and we want to use the smaller array multiple times to perform some operation on the larger array.

```python
x = np.array([[1,2], [3,4], [5,6]])
print('x='); print(x)
print('')
y = np.array([2, 3])
print('y='); print(y)
print('')
print('x + y='); print(x + y)
```

## Linear Regression with Single Neuron (線性迴歸)

### 產生隨機的資料

```python
import random
import numpy as np
import matplotlib.pyplot as plt

xs = []; ys = []

for idx in range(50):
  x = random.uniform(0, 10); y = x * 0.5 + 0.3 + random.uniform(-0.2, 0.2) #noise
  xs.append(x); ys.append(y)

#將x和y的資料製作成numpy array
data = np.array(xs, dtype=np.float32)
ground_truth = np.array(ys, dtype=np.float32)

plt.rcParams["figure.figsize"] = (8, 7)
plt.plot(xs, ys, 'ro')
plt.title(f'Regression line: f(x) = 0.5x + 0.3 + noise')
plt.xlabel('x'); plt.ylabel('f(x)')
plt.show()
```

### 建立模型

```python
import torch
import torch.nn as nn
from torch.optim import optimizer

print(f'Shape of data: {data.shape}, Type of data: {type(data)}') #印出data(x點座標)的datatype
print(f'Shape of ground truth: {ground_truth.shape}, Type of ground truth: {type(ground_truth)}') #印出ground_truth(y點座標)的datatype

#將資料(x,y)從numpy array轉換成tensor
#unsqueeze的功能是對資料的維度進行壓縮
data = torch.from_numpy(data).unsqueeze(1) #unsqueeze(1)的目的是壓縮一維的資料
ground_truth = torch.from_numpy(ground_truth).unsqueeze(1)
print(f'Shape of data: {data.shape}, Type of data: {type(data)}')
print(f'Shape of ground truth: {ground_truth.shape}, Type of ground truth: {type(ground_truth)}')

#Model
class LinearRegression(nn.Module):
  #建構式
  def __init__(self, input_dim, output_dim):
    super(LinearRegression, self).__init__()
    self.learner = nn.Linear(input_dim, output_dim)
    nn.init.zeros_(self.learner.weight); nn.init.zeros_(self.learner.bias)
  #forwarding
  def forward(self, x):
    return self.learner(x)
  def get_params(self):
    #detach()阻斷反向傳播，numpy()將tensor轉成numpy
    return self.learner.weight.detach().numpy()[0][0], self.learner.bias.detach().numpy()[0]
```

### 跑結果

```python
model = LinearRegression(1, 1) #建立Model
criterion = nn.MSELoss() #選擇loss function
optimizer = torch.optim.SGD(model.parameters(), lr=5e-4) #選擇optimizer
weights = []; biases = []; losses = [] #存權重、偏移和損失
max_loss = -1

epochs = 100 #要進行的訓練次數
xt = [1, 2, 5, 10, 50, epochs] #要印出結果的回合
for epoch in range(epochs):
  loss = 0 #每個epoch將loss歸0
  
  #算loss及做訓練
  for idx in range(len(data)):
    pred = model(data[idx]) #算出預測值
    loss += criterion(pred, ground_truth[idx]) #算loss function(用y->預測值和y head相減->正確答案)
  loss.backward() #做backward
  optimizer.step() #當做完backward後這行能更新所有參數
  optimizer.zero_grad() #將梯度歸0

  #當epoch是指定的回合數，印出結果
  if epoch+1 in xt:
    w, b = model.get_params()
    weights.append(w); biases.append(b); losses.append(loss.detach().numpy())

    #畫作標點的圖
    f, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
    ax1.plot(xs, ys, 'ro') #ro代表畫紅點
    mx = [-2, 12]
    my = [mx[0]*w+b, mx[1]*w+b] #用訓練的w來算出結果
    ax1.plot(mx, my, 'k-') #畫線性回歸的線
    ax1.set_xlim([-0.5, 10.5]); ax1.set_ylim([-0.5, 6.5]) #xlim和ylim是用來限制x軸和y軸的範圍
    ax1.set_xlabel('x'); ax1.set_ylabel('y')

    
    if b < 0: #bias小於0
      ax1.set_title(f'Regression line: y={"{0:.4f}".format(w)}x{"{0:.4f}".format(b)}')
    else: #bias大於等於0
      ax1.set_title(f'Regression line: y={"{0:.4f}".format(w)}x+{"{0:.4f}".format(b)}')
    
    if max_loss == -1:
      max_loss = loss.detach().numpy()

    #畫loss圖
    ax2.plot(range(1, len(losses)+1), losses, '-o') #畫loss的藍點
    plt.xticks(range(1, 7), xt) #將點連線
    ax2.set_xlim([0.5, 6.5]); ax2.set_ylim([-2.5, max_loss+2.5]) #設定x軸和y軸的範圍
    ax2.set_title(f'Losses (epoch {epoch+1} loss: {"{0:.4f}".format(loss)})') #設定標題
    ax2.set_xlabel('epoch') #設定x軸文字
    print(f'Epoch: {epoch+1}, Loss = {loss.item(): .4f}, Model parameters = [w: {w}, b: {b}]')
    plt.show()
```

## Binary Classification with Single Neuron (二元分類)

### 產生隨機的資料

```python
import numpy as np
import random
from matplotlib import pyplot as plt

x1s = []; x2s = []; ys = [] #點座標有 x1, x2, y(代表分類)
ax1s = []; ax2s = [] # (x1, x2) in class a, labeled as 0 -> 類別a
bx1s = []; bx2s = [] # (x1, x2) in class b, labeled as 1 -> 類別b

for idx in range(100):
  x1 = random.uniform(0, 10); x2 = random.uniform(0, 10) # (x1,x2)皆為亂數
  x1s.append(x1); x2s.append(x2)
  if x1 * 0.4 + 2.5 - x2 < 0: #若x1和x2產生的y值小於0，則被分類至class a
    ax1s.append(x1); ax2s.append(x2)
    ys.append(0)
  else: #若x1和x2產生的y值>=0，則被分類至class b
    bx1s.append(x1); bx2s.append(x2)
    ys.append(1)

data = np.dstack((np.array(x1s, dtype=np.float32), np.array(x2s, dtype=np.float32))) #dstack是用深度順序去堆疊arrays
ground_truth = np.array(ys, dtype=np.float32)

plt.rcParams["figure.figsize"] = (8, 7)
plt.plot(ax1s, ax2s, 'ro') #被分類為a的為紅點
plt.plot(bx1s, bx2s, 'bo') #被分類為b的為藍點
plt.xlabel('x1')
plt.ylabel('x2')
plt.title(f'Classifier: x2 = 0.4x1 + 2.5')
plt.show()
```

### 建立模型

```python
import torch
import torch.nn as nn
from torch.optim import optimizer

print(f'Shape of data: {data.shape}, Type of data: {type(data)}')
print(f'Shape of ground truth: {ground_truth.shape}, Type of ground truth: {type(ground_truth)}')

#將numpy array轉換為tensor
data = torch.from_numpy(data).permute(1, 2, 0).squeeze(-1)
ground_truth = torch.from_numpy(ground_truth).unsqueeze(1)
print(f'Shape of data: {data.shape}, Type of data: {type(data)}')
print(f'Shape of ground truth: {ground_truth.shape}, Type of ground truth: {type(ground_truth)}')

class BinaryClassification(nn.Module):
  def __init__(self, input_dim, output_dim):
    super(BinaryClassification, self).__init__()
    self.learner = nn.Linear(input_dim, output_dim)
    nn.init.zeros_(self.learner.weight); nn.init.zeros_(self.learner.bias)
  def forward(self, x):
    return torch.sigmoid(self.learner(x)) #使用sigmoid作為激勵函數
  def get_params(self):
    return self.learner.weight.detach().numpy()[0][0], self.learner.weight.detach().numpy()[0][1], self.learner.bias.detach().numpy()[0]
```

### 跑結果

```python
model = BinaryClassification(2, 1) #input dim: 2, output dim: 1
criterion = nn.BCELoss() #使用Binary Cross Entropy作為loss function
optimizer = torch.optim.SGD(model.parameters(), lr=2e-3) #設定optimizer
weights = []; biases = []; losses = []
max_loss = -1

epochs = 1000 #要進行的訓練次數
xt = [1, 2, 10, 100, 500, epochs] #要印出結果的回合
for epoch in range(epochs):

  #算loss及做訓練
  loss = 0 #每個epoch將loss歸0
  for idx in range(len(data)):
    pred = model(data[idx]) #算出預測值
    loss += criterion(pred, ground_truth[idx]) #算loss function(用y->預測值和y head相減->正確答案)
  loss.backward() #做backward
  optimizer.step() #當做完backward後這行能更新所有參數
  optimizer.zero_grad() #將梯度歸0
  
  #當epoch是指定的回合數，印出結果
  if epoch+1 in xt:
    w1, w2, b = model.get_params()
    w = -w1/w2; b = -b/w2
    weights.append(w); biases.append(b); losses.append(loss.detach().numpy())

    #畫作標點的圖
    f, (sub1, sub2) = plt.subplots(1, 2, figsize=(10, 5))
    sub1.plot(ax1s, ax2s, 'ro') #class a的為紅點
    sub1.plot(bx1s, bx2s, 'bo') #class b的為藍點
    mx1 = [-2, 12]
    mx2 = [mx1[0]*w+b, mx1[1]*w+b] #用訓練的w來算出結果
    sub1.plot(mx1, mx2, 'k-') #畫二元分類的線
    sub1.set_xlim([-0.5, 10.5]); sub1.set_ylim([-0.5, 10.5]) #設定x軸和y軸的範圍
    sub1.set_xlabel('x1'); sub1.set_ylabel('x2') #設定x軸和y軸的文字


    if b < 0: #bias小於0
      sub1.set_title(f'Classifier: x2={"{0:.4f}".format(w)}x1{"{0:.4f}".format(b)}')
    else: #bias大於等於0
      sub1.set_title(f'Classifier: x2={"{0:.4f}".format(w)}x1+{"{0:.4f}".format(b)}')
    
    if max_loss == -1:
      max_loss = loss.detach().numpy()

    #畫loss的圖
    sub2.plot(range(1, len(losses)+1), losses, '-o') #畫loss的藍點
    plt.xticks(range(1, 7), xt) #將點連線
    sub2.set_xlim([0.5, 6.5]); sub2.set_ylim([-2.5, max_loss+2.5]) #設定x軸和y軸的範圍
    sub2.set_title(f'Losses (epoch {epoch+1} loss: {"{0:.4f}".format(loss)})') #設定標題
    sub2.set_xlabel('epoch') #設定x軸的文字
    print(f'Epoch: {epoch+1}, Loss = {loss.item(): .4f}, Model parameters = [w: {w}, b: {b}]')
    plt.show()
```
